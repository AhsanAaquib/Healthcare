1. Applying Logistic Regression

```python
features = healthcare.iloc[:,[0,1,2,3,4,5,6,7]].values
label = healthcare.iloc[:,8].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train) 
```

```python
print(logreg.score(X_train,y_train))
print(logreg.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(label,logreg.predict(features))
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(label,logreg.predict(features)))
```

```python
#Preparing ROC Curve (Receiver Operating Characteristics Curve)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn import metrics
# Predict Probabilities
probs = logreg.predict_proba(features)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# Calculating AUC
auc = roc_auc_score(label, probs)
print('AUC: %.3f' % auc)
print('Accuracy: ',metrics.accuracy_score(label,logreg.predict(features)))
# Calculating roc curve
fpr, tpr, thresholds = roc_curve(label, probs)
# Plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# Plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
```

```python
cm=pd.crosstab(label,logreg.predict(features),rownames=['ACtual'],colnames=['Predicted'])
sns.heatmap(cm,annot=True)
```

### Comoparing precision recall curve for various models 

```python
#Precision Recall Curve for Logistic Regression
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import average_precision_score
# Predict Probabilities
probs = logreg.predict_proba(features)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# Predict class values
yhat = logreg.predict(features)
# Calculating precision-recall curve
precision, recall, thresholds = precision_recall_curve(label, probs)
# Calculating F1 score
f1 = f1_score(label, yhat)
# Calculating precision-recall AUC
auc = auc(recall, precision)
# Calculating average precision score
ap = average_precision_score(label, probs)
print('f1=%.3f auc=%.3f ap=%.3f' % (f1, auc, ap))
# Plot no skill
plt.plot([0, 1], [0.5, 0.5], linestyle='--')
# Plot the precision-recall curve for the model
plt.plot(recall, precision, marker='.')
```
